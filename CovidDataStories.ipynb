```python
from bs4 import BeautifulSoup as bs
import requests as req
import pandas as pd
import datetime
import time
import csv
```

```python
url="https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
```

```python

# page here is just a request object
#This establishes connection to the website
page= req.get(url)
```

```python
#soup is an object of bs class. It allows us to interact with html DOM now
soup = bs(page.text, "html.parser")
```

```python
print(soup.title.get_text())
#Alternatively, you could write
print(soup.title.string)
```

```python
'''All tables in the particular web page'''

AllTables = soup.find_all('table')
```

```python
# There are 70 tables here in total 
print(len(AllTables))
```

```python

# Now the task at hand is filter out a particular table that you have in sight
# Since that particular has caption in it, you should go ahead and 
# remove all those tables which are caption-less

filteredTable = [table for table in AllTables if table.caption is not None ]
```

```python
print(len(filteredTable))
```

```python
# We wish to what all tables got selected in the filteredTables list

l=[]
for table in filteredTable:
  l.append(table.caption.get_text())
  
```

```python
print(l)
```

```python
requiredTable = None
for table in filteredTable:
  if table.caption.get_text() == 'COVID-19 cases, deaths, and rates by location[5]\n':
    requiredTable= table
    break
```

```python
print(requiredTable)
```

```python
header_tags = requiredTable.find_all('th')
```

```python
print(header_tags)
```

```python
headers = [str(header.text.strip()) for header in header_tags]
```

```python
print(headers)
```

```python
AllData = requiredTable.find_all('tr')
new_headers=headers[1:5]
print(new_headers)
```

```python
col1 = headers[5:]
print(len(col1))
```

```python
rows = []
for row in AllData:
  value = row.find_all('td')
  beautified_value = [dp.text.strip() for dp in value]
  print(beautified_value)
  rows.append(beautified_value)
```

```python
rows= rows[1:]
```

```python
print(rows)
j=5
for row in range(int(len(rows))):
  for i in range(int(len(row))):
    if i==0:
      rows[row][i]=headers[j]
      j=j+1
```

```python
with open('Covid Data Stats.csv','w',encoding='UTF8') as output:
  writer= csv.writer(output)
  writer.writerow(new_headers)
  writer.writerows(rows)
```
